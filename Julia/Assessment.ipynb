```julia
#Function return sum of arguments passed
#single line function
sumArg(x,y) = x+y
println(sumArg(1,2))
```

```julia
#Function returns product of arguments
function prodArg(l,m)
    return l*m
end
println(prodArg(2,3))
```

```julia
#For loop displays values of x and y in -10 to 10 range
f(x) = 2 * x^2 - 9
for x = -10:10    
  println("(",x, ", ", f(x), ")")
end
```

```julia
#Lets display the graph 
using Plots
gr()
x = collect(-10:10)
f(x) = 2 * x^2 - 9

plot(x,f.(x),marker = :hex, leg=true, linewidth = 2, linecolor=:red,title="plot the Graph")
```
